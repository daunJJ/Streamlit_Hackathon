```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df_info = pd.read_csv('/content/drive/MyDrive/Streamlit 해커톤/data/info.csv')
df_info.head(3)
```

```python
df_name = pd.read_csv('/content/drive/MyDrive/Streamlit 해커톤/data/crawling_link.csv')
df_name.head(3)
```

```python
# 데이터를 합치기 위해 중복 열 제거
df_name = df_name.drop(['index', 'link'], axis=1)
```

```python
# 두 데이터 합치기
df = pd.concat([df_name, df_info], axis=1)
df
```

```python
# 열 순서 변경
df = df[['index', 'title', 'category', 'body', 'image', 'link']]
df
```

```python
# 상품명에 포함된 '_'를 ' '으로 바꾸기
df['title'] = df['title'].str.replace('_', ' ')
df
```

```python
df['title'].value_counts()
```

```python
# 중복 제거
df = df.drop_duplicates(subset='title', keep='first').reset_index(drop=True)
df.shape
```
